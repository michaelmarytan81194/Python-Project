```python
#1. Import necessary libraries
import glob
import pandas as pd
import json
from datetime import datetime
```

```python
#2. Create extract functions for each file:
def extract_from_json (input_file):
    df = pd.read_json(input_file)
    return df

def extract_from_csv (input_file):
    df = pd.read_csv(input_file, index_col = 0)
    return df
```

```python
# Examine how data looks
extract_from_csv('exchange_rates.csv')
```

```python
extract_from_json('bank_market_cap_1.json')
```

```python
def extract():
    df = pd.DataFrame(columns=['Name', 'Market Cap (US$ Billion)'])
    for jsonfile in glob.glob('*.json'):
        extracted_data = df.append(extract_from_json (jsonfile), ignore_index = True)
    return extracted_data
```

```python
data = extract()
data
```

```python
#3. Tranform function:
def transform (input_data): #input_data is a DataFrame
    input_data['Market Cap (US$ Billion)'] = round(input_data.loc[:,'Market Cap (US$ Billion)'] * 0.79 ,3) 
    input_data.rename(columns={"Market Cap (US$ Billion)": "Market Cap (GBP$ Billion)"}, inplace = True) #inplace Ä‘á»ƒ thay Ä‘á»•i vÃ o dataframe
    return input_data
```

```python
transform_data = transform(data)
transform_data
```

```python
def log(message):
    timestamp_format = '%Y-%m-%d-%H:%M:%S'
    now = datetime.now() #retrieve current time
    timestamp = now.strftime(timestamp_format)
    with open ('logfile.txt', 'a') as f:
        f.write('At ' + timestamp + ' '  + message +'\n')
```
