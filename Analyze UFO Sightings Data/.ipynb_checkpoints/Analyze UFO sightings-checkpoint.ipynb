```python
#1. Import necessary libraries
import csv
```

```python
#2. Extract data:
ufosightings = []
with open('ufo-sightings.csv', 'r') as handle_file:
    file_reader = csv.DictReader(handle_file)
    for row in file_reader:
        ufosightings.append(row)
```

```python
#Preview the extracted_data
ufosightings
```

```python
#3.Analysis phase
#3a. Total of sightings
ufosightings_count = 0
for sight in ufosightings:
    ufosightings_count +=1
print(ufosightings_count)
```

```python
#3b. Sightings in the US
sightings_us = [row for row in ufosightings if row['country'] == 'us']
print(len(sightings_us))

```

```python
#3c. Find the "fireball" sighting(s) that lasted more than ten seconds in US
#1. Function for checking validation
def is_valid_duration(duration_as_string):
    try:
        duration = float(duration_as_string)
    except ValueError:
        return False
    else:
        return duration > 10
#2. List of result:
fball = [row for row in sightings_us if is_valid_duration(row['duration (seconds)']) and row['shape'] == 'fireball']
fball
```

```python
#3d. Sort the above list by duration. Find the datetime and duration of the longest sighting
fballsorted = sorted(fball, key = lambda x: float(x['duration (seconds)']), reverse = True)[0]
print([fballsorted['datetime'], fballsorted['duration (seconds)']])
```

```python
#3e. Get the state
fballsorted['state']
```

```python
#3f.Find the min duration
def valid_above_zero(duration_as_string):
    try:
        duration = float(duration_as_string)
    except ValueError:
        return False
    else:
        return duration 
    
filtered = [row for row in ufosightings if valid_above_zero(row['duration (seconds)']) != 0]
min_duration = min(filtered, key = lambda x: valid_above_zero(x['duration (seconds)']))
min_duration
```

```python
#3g. What are the top 3 shapes sighted, and how many sightings were there for each
sightings_shapes= [row['shape'] for row in ufosightings ]
sightings_shapes
```

```python
count = {}
for shape in sightings_shapes:
    count[shape] = count.get(shape, 0) + 1
count
```

```python
lst_count_shape = []
for k, v in count.items():
   lst_count_shape.append([k,v])
lst_count_shape
```

```python
top3shapes = sorted(lst_count_shape, key = lambda x: float(x[1]), reverse = True)[:3]
top3shapes
```
