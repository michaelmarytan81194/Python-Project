```python
import pandas as pd
xls_data = pd.ExcelFile('AD_Bonus_08.2022-08.2023_Final.xlsx')
xls_structure = pd.ExcelFile('AD Structure.xlsx')
```

```python
data = xls_data.parse('Data')
data
```

```python
structure = xls_structure.parse('AD Structure')
structure
```

```python
join = pd.merge(left = structure, right = data, how ='inner', left_on = 'AD Code', right_on = 'AD Code')
join
```

```python
territory = data['Territories'].drop_duplicates()
lst_territory = []
for item in territory:
    lst_territory.append(item)
final_territory = lst_territory[:12]
```

```python
#South 3
south_3 = join[join['Territories_y'] == 'South 3']
with pd.ExcelWriter('Result.xlsx', mode='a') as writer:  
    south_3.to_excel(writer, sheet_name='South 3')
```

```python
columns = []
for column in join.columns:
    columns.append(column)
print(columns)
```

```python
final_columns = ['Territories_x', 'Code TD', 'Name TD', 'Code TAD', 'Name TAD', 'Code SRD', 'Name SRD', 'Code RD', 'Name RD', 'Code RAD', 'Name RAD', 'Code SZD', 'Name SZD', 'AD Code', 'AD Name_x', 'Grade', 'Date Appointed', 'Date Terminated', 'Status', 'Terdate', 'Date_Appointed', 202208, 202209, 202210, 202211, 202212, 202301, 202302, 202303, 202304, 202305, 202306, 202307, 202308, 'Grand Total', 'Note Holding']
for territory in final_territory:
    mien = join[join['Territories_y'] == territory]
    check_mien = mien[final_columns].drop_duplicates()
    with pd.ExcelWriter('Result1.xlsx', mode='a') as writer:  
        check_mien.to_excel(writer, sheet_name=territory, index = False)
```
