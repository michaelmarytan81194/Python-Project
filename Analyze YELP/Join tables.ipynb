```python
#1. Necessary libraries
import pandas as pd
```

```python
#2. Extract and load data
xls = pd.ExcelFile('yelp.xlsx') #handle
df = xls.parse('yelp_data')
df.head()
```

```python
df_cities = xls.parse('cities')
df_cities.head()
```

```python
df_states = xls.parse('states')
df_states
```

```python
#3a. SELECT * FROM DF a INNER JOIN DF_CITIES b ON a.city_id = b.id
df_join_1 = pd.merge(left=df, right = df_cities, how ='inner', left_on='city_id', right_on= 'id')
df_join_1
```

```python
#3b.  Join the state table
df_join_2 = pd.merge(left=df_join_1, right = df_states, how ='inner', left_on = 'state_id', right_on = 'id')
df_join_2
```

```python
atts = ['name', 'city', 'state']
df_join_2[atts].head(100)
```

```python
#4. Remove duplicate columns id_x, id_y
del df_join_2['id_x']
del df_join_2['id_y']
```
