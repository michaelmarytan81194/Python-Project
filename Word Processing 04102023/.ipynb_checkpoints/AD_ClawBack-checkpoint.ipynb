```python
handle_excel = pd.ExcelFile("Excel.xlsx")
data = handle_excel.parse('Sheet1')
data.head(5)
```

```python
#Xóa dấu tiếng Việt để đưa vào dòng chuyển khoản ngân hàng
BANG_XOA_DAU_FULL = str.maketrans(
    "ÁÀẢÃẠĂẮẰẲẴẶÂẤẦẨẪẬĐÈÉẺẼẸÊẾỀỂỄỆÍÌỈĨỊÓÒỎÕỌÔỐỒỔỖỘƠỚỜỞỠỢÚÙỦŨỤƯỨỪỬỮỰÝỲỶỸỴáàảãạăắằẳẵặâấầẩẫậđèéẻẽẹêếềểễệíìỉĩịóòỏõọôốồổỗộơớờởỡợúùủũụưứừửữựýỳỷỹỵ",
    "A"*17 + "D" + "E"*11 + "I"*5 + "O"*17 + "U"*11 + "Y"*5 + "a"*17 + "d" + "e"*11 + "i"*5 + "o"*17 + "u"*11 + "y"*5,
    chr(774) + chr(770) + chr(795) + chr(769) + chr(768) + chr(777) + chr(771) + chr(803) # 8 kí tự dấu dưới dạng unicode chuẩn D
)

def xoa_dau_full(txt: str) -> str:
    return txt.translate(BANG_XOA_DAU_FULL)
```

```python
#Phải run cái này trước để tránh mất format của cột Amount ở kiểu number
data['Amount_Word'] = pd.Series([n2w(str(number)).capitalize() for number in data['Paid']])
#làm tròn số
data['Paid'] = pd.Series([int(number) for number in data['Paid']])
data['Holding'] = pd.Series([int(number) for number in data['Holding']])
data['Total'] = pd.Series([int(number) for number in data['Total']])
data
```

```python
#ỏ đây đã thành string
data['Paid'] = pd.Series([f'{number:,}'.replace(",",".") for number in data['Paid']])
data['Holding'] = pd.Series([f'{number:,}'.replace(",",".") for number in data['Holding']])
data['Total'] = pd.Series([f'{number:,}'.replace(",",".") for number in data['Total']])
```

```python
data['Amount_Word'] = pd.Series([word.replace("lẽ","lẻ") for word in data['Amount_Word']])
```

```python
data
```

```python
for index, row in data.iterrows():
    if row['Holding'] != '0':
        context = { 
            'Code': row['Code'],
            'Gender' : row['Gender'],
            'AD_Name' : row['AD_Name'],
            'AD_Name_Eng': xoa_dau_full(row['AD_Name']),
            'Add': row['Add'],
            'Grade': row['Grade'],
            'Grade2': row['Grade2'],
            'date': row['Day'],
            'month': row['Month'],
            'year': row['Year'],
            'Contract_No': row['Contract_No'],
            'Contract_Date': row['Contract_Date'],
            'Paid' : row['Paid'],
            'Holding': row['Holding'],
            'Total': row['Total'],
            'Amount_Word': row['Amount_Word'],
            'gender': row['gender']
                    }
        doc_2.render(context)
        doc_2.save(f"result/{row['Grade']}-{row['AD_Name']}.docx")
    else:
        context = { 
            'Code': row['Code'],
            'Gender' : row['Gender'],
            'AD_Name' : row['AD_Name'],
            'AD_Name_Eng': xoa_dau_full(row['AD_Name']),
            'Add': row['Add'],
            'Grade': row['Grade'],
            'Grade2': row['Grade2'],
            'date': row['Day'],
            'month': row['Month'],
            'year': row['Year'],
            'Contract_No': row['Contract_No'],
            'Contract_Date': row['Contract_Date'],
            'Paid' : row['Paid'],
            'Amount_Word': row['Amount_Word'],
            'gender': row['gender']
                    }
        doc_1.render(context)
        doc_1.save(f"result/{row['Grade']}-{row['AD_Name']}.docx")
```
