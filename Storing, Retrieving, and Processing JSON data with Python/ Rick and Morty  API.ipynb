```python
#. Build the class for characters
class Character:
    def __init__(self, name, gender, species, origin, status, image, number_of_episodes):
        self.name = name
        self.gender = gender
        self.species = species
        self.origin = origin
        self.status = status
        self.image_url = image
        self.number_episodes = number_of_episodes
    def show_character(self):
        print(self.name, self.gender)
```

```python
#3. Build Load function:
def load_data():
    #a. We need to have a list of result:
    url = 'https://rickandmortyapi.com/api/character/?page=1'
    response = requests.get(url)
    json_res = response.json()
    json_res_results = json_res['results']

    #b. Create an empty list for characters:
    characters= []

    #c. Iterate:
    for obj in json_res_results: #obj is a dict
        name = obj['name']
        gender = obj['gender']
        species = obj['species']
        origin = obj['origin']['name']
        status = obj['status']
        image_url = obj['image']
        number_episodes = len(obj['episode'])
        #d. Create an instance of character class
        character = Character(name, gender, species, origin, status, image_url, number_episodes) #NOTE11
        character.show_character() #show summary info of character
        characters.append(character) #NOTE11
    return characters

    
```
