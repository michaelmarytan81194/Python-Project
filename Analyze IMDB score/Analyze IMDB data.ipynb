```python
#1. Import file and library
import pandas as pd
xls_handle = pd.ExcelFile('imdb.xlsx')
df = xls_handle.parse('imdb')
df_directors = xls_handle.parse('directors')
df_countries = xls_handle.parse('countries')
```

```python
df.head()
```

```python
df_directors.head()
```

```python
df_countries.head()
```

```python
df = pd.merge(left=df, right=df_directors, how ='inner', left_on = 'director_id', right_on = 'id')
df
```

```python
df = pd.merge(left = df, right = df_countries, how ='inner', left_on = 'country_id', right_on = 'id')
df
```

```python
#2. Clean the redundant character at the end of movie title
df['movie_title'] = df['movie_title'].str.replace("Ê","")
df
```
